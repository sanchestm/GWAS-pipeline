```python
import sys
sys.path.append('../')
from gwas_class_auto import *
def get_trait_descriptions_f( traits):
    qq =  pd.read_csv(f'data_dict_ny_derek_water_staging.csv')
    out = []
    for trait in traits:
        try: 
            out +=  [qq[qq.measure == trait.replace('regressedlr_', '')].description.iloc[0]]
        except:
            out +=  ['UNK']
    return out

df = pd.read_csv('food_fluid as of 20181029.csv', usecols=range(17), dtype = {'rfid': str}).drop_duplicates(subset = 'rfid').set_index('rfid')
df['sex'] = df.food_fluid_cov_sex
df['food_fluid_cov_age'] = df.food_fluid_cov_age.abs()
df.loc[['00077E7BFE', '00077E9902', '00077E7C36'], 'food_fluid_trait_water_intake_ml'] = np.nan
df.loc[['00077E7C85'], 'food_fluid_trait_bursts'] = np.nan 
df.loc[['00077E8332', '00077E9BF0', '00077E88EB', '00077E9C37'], 'food_fluid_trait_food_intake_g'] = np.nan
df = df.reset_index()
df.to_csv('raw_data.csv')

traits_ = df.columns[df.columns.str.startswith('food')][3:].to_list()
traits_d = get_trait_descriptions_f( traits_)
gwas = gwas_pipe(path = f'',
             all_genotypes = '/projects/ps-palmer/tsanches/gwaspipeline/gwas/zzplink_genotypes/round10', #'round9_1.vcf.gz',
             data = df,
             project_name = 'ny_derek_water_staging',
             traits = traits_ ,
             trait_descriptions= traits_d,
             threads = 8)
```
